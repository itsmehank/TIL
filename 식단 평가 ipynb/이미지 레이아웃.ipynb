```python
from math import pi
import numpy as np
import pandas as pd
from bokeh.io import output_file, show
from bokeh.palettes import Category20c
from bokeh.plotting import figure
from bokeh.transform import cumsum
from bokeh.models import LabelSet, ColumnDataSource
from bokeh.models import HoverTool



# output_file("pie.html")






# 표현하고자 하는 데이터 만들기
numbers = [90,72,59]
nutrition = ["탄수화물","단백질","지방"]







# 표현할 데이터를 dictionary 형태로 만들어주고 이후 ColumnDataSource 객체로 만들어서 그래프 그릴때 source 인자로 넣어주고 
# 해당 코드에서는 딕셔너리의 키값으로 각각의 데이터를 넣어준다.
# wedge의 경우 각도를 이용해서 그래프를 그리기에 각 데이터의 비율 * 2 * pi를 해줘야 한다.(pi는 math 라이브러리에서 불러온다.)

data1 = {'nutrition': nutrition,

'value': numbers,

'angle': [n/sum(numbers)*2*pi for n in numbers],

'percentage': [d/sum(numbers)*100 for d in numbers],
         
'cumulative_angle':[(sum(numbers[0:i+1])- (item/2))/sum(numbers)*2*pi for i,item in enumerate(numbers)],

'color': ['#FA3838','#FFC300','#FF5733'] }






# wedge 그래프의 각 영역 위에 라벨을 붙여주기 위한 데이터이다. 라벨은 직접 좌표값을 지정해서 붙여줘야 하는데 
# 이 좌표값을 구하기 위해 일단 그래프를 그릴때 grid와 axis를 살려서 그려준 다음 하나하나 각 영역의 좌표를 눈으로 확인해서
"""
그래프 출력단 바로 전에 아래의 코드 주석처리해서 출력하여 좌표값을 구한 다음 나중에 실제 출력시에는 원상복구한다.
# p.axis.axis_label=None
# p.axis.visible=False
# p.grid.grid_line_color = None
"""
# data["x"], data["y"]에 넣어준다. 이후 이 좌표를 활용해서 해당 영역에 라벨을 출력하는 코드를 작성할 것이다.
# 라벨을 붙이기 위해 위의 좌표정보와 함께 label 정보도 필요한데 이 역시 data["label"]에 저장을 해줘야 한다.

data1['label'] = [(nutri +" "+ format(percen, '.2f') + "%") for nutri,percen in zip(data1['nutrition'],data1['percentage'])]
data1['x']= np.cos(data1['cumulative_angle'])*0.5
data1['y']= np.sin(data1['cumulative_angle'])*0.4






# 데이터의 입력이 끝나면 해당 dictionary 객체를 이용해 ColumnDataSource 객체를 만들어준다.(보통 source로 할당한다)
source1 = ColumnDataSource(data=data1)






# 툴팁은 파이차트의 각 영역위에 마우스를 올려놓을 때 나타나는 창이다. 리스트 형식을 사용해 창위에 표현될 인자들을 
# 튜플형태로 넣어준다. 튜플 = ("해당값의 닉네임(아무거나 해도된다_출력되는게 아님)","@<data의 key값>")
# 이때 숫자의 경우 표현 형식을 맞춰주려면 @<data의 key값>{0.2f}%등으로 표현 형식을 정해준다.
# TOOLTIPS = [
# ("울랄라", "@numbers"),
# ("Percentage", "@percentage{0.2f}%")
# ]


# 위의 Tooltip이 아닌 좀 더 customize된 tooltip을 사용하기 위해(폰트 사이즈 변경이 목적) 새로 작성


TOOLTIPS1 = """
    <div>
        <div>
            <span style="font-size: 14px; font-weight: bold;">@nutrition : @value g </span>
        </div>
    </div>
"""









# firuge 객체를 만든다 = 도화지를 그려준다.
# 각 인자에 대한 설명
'''
plot_height : (이후 그려질)그래프의 출력 높이를 설정한다.
plot_width : (이후 그려질)그래프의 출력 너비를 설정한다.
>>> 위의 height와 width는 그래프의 실제 크기가 아니라 화면에 출력되는 크기다. 다시말하면 이후 wedge 메소드를 이용해서
>>> 그릴 그래프의 경우 x좌표, y좌표의 값과 반지름을 이용해 실제 크기를 설정해주는데 
>>> 이 실제 크기가 화면에 표시될때는 화면에서 x축 위의 0.1cm 가 어느정도의 크기인지 정하는 기준에 따라 그래프의 크기가 달라질 수 있다.
>>> 따라서 그래프의 출력크기는 그래프의 실제 좌표상 크기와 상관없이 따로 설정하는 것이 가능하고 이를 위한 변수가 바로 위 두개다. 
title : figure의 타이틀로 그래프 오른쪽 상단에 작게 적히게 된다.
toolbar_location : 도구창의 위치로 도구창을 띄우기 싫다면 None 값을 주면 된다.
tools : 도구창에 표현할 도구들을 설정해주는 변수로 아래에서 사용하는 hover는 그래프를 고정시켜주는 기능을한다(마우스로 영역이동 안됨)
tooltips : tooltips는 그래프의 해당 영역에 마우스를 올리면 정보창이 뜨는 옵션이다. 
>>> tooltips는 옵션값으로 "@<data의 key값>: @<data의 key값>" 형식으로 입력한다. 
x_range : 그래프가 표현될 x축 영역을 정의해준다.
'''
p1 = figure(plot_height=600,plot_width=600, title="기본 영양소 정보", toolbar_location=None,
            tools="hover",tooltips=TOOLTIPS1, x_range=(-2, 2))







# wedge는 bokeh에서 제공하는 파이차트이다. 이 차트는 각도를 이용해서 각 영역의 크기를 구해서 그려준다.
# wedge 메소드의 인자로는 
'''
x = 해당 그래프의 중심 x 좌표
y = 해당 그래프의 중심 y 좌표
radius = 해당 그래프의 반지름 크기
start_angle = data안의 각도값의 키값인 angle을 cumsum()안에 넣어주고 include_zero = True 옵션을 준다.
end_angle = data안의 각도값의 키값인 angle을 cumsum()안에 넣어서 입력한다.
line_color = 각 영역사이의 구분선의 색을 정하는 것이다.(white가 깔끔한다)
line_width = 각 영역 사이의 줄의 굵기를 정해준다.
fill_color = 각 영역의 color를 list형식으로 입력해준다.
legend = 범례로 사용할 이름들의 리스트를 입력해준다. (angle데이터와 순차적으로 매칭된다.)
source = ColumnDataSource 객체를 입력값으로 넣어준다.
'''
# 등이 있다.
p1.wedge(x=0, y=0, radius=1.5,
        
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        
        line_color="white",line_width = 3.5, fill_color='color', legend='nutrition', source=source1)







# 파이차트 위에 표현할 라벨을 만들고 해당 라벨을 figure객체에 add해줄 것이다.
# 기본적으로 라벨은 좌표값으로 입력되는 것이기에 만들어진 파이차트 그림을 x축 y축과 함께 보면서 좌표값을 직접
# 설정해주는 작업이 필요하다.
# 1. LabelSet() 객체를 만들어준다.
'''



'''
# 2. 해당 LabelSet 객체를 figure 객체에 add_layout 메소드를 사용해서 입혀준다.
# Labelset의 인자들
'''
x = 각 라벨의 x좌표 값
y = 각 라벨의 y좌표 값
text = 각 라벨의 값
y_offset = y 값의 시작점을 설정하는 것으로 값으로 입력한 숫자는 실제 y축의 단위가 아닌 byte의 단위로 훨씬 더 작다.
>>> offset이란 해당 데이터에서 실제로 우리가 원하는 값이 몇 byte뒤에 있는지에 대한 변수이다. 
>>> 데이터 앞부분이 이상데이터 혹은 관련없는 내용의 데이터로 corrupt되어 있다면 offset을 사용해서 데이터를 읽어 올 
>>> 시작위치를 뒤로 땡겨줄 수 있다.
text_font_size = 위에서 정해준 text의 폰트 사이즈를 정해준다
text_color = text의 컬러를 정해준다.
source = 이 LabelSet 객체를 만들기 위해 사용한 데이터를 담고있는 ColumnDataSource 객체를 입력값으로 넣어준다.
text_align = 영역위에 표현될 텍스트의 정렬 방법을 알려준다. 
'''
labels1 = LabelSet(x='x', y='y', text="label", y_offset= 0,
text_font_size="14pt", text_color="black",
source=source1, text_align='center')

p1.add_layout(labels1)


# 기타사항 1. data['angle'] = data['value']/data['value'].sum() * 2*pi
# >>> wedge를 그리기 위해서는 angle이 필요하다. 그릴때만 사용한다.





# 기타사항 2. Category20c란? _ 갯수별 컬러코드를 담고있는 데이터
'''
Category20c는 3개부터 20개까지의 컬러코드를 인자로 하는 dictionary이다. 데이터의 수에 맞게 컬러 리스트를 불러올 수 있다.
(예시)
Category20c(3) >>>>  3개의 색상을 담고있는 리스트를 불러온다.
'''





# 그래프의 축과 격자 없이 출력하기 위한 코드이다.
p1.axis.axis_label=None
p1.axis.visible=False
p1.grid.grid_line_color = None

show(p1)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import BoxAnnotation
from bokeh.models import ColumnDataSource, Range1d, LabelSet, Label,LegendItem,Legend

day2 = ["월","화","수","목","금","토","일"]
cosine_simil2 = [99,98,97,96,95,99,98]
sodium2 = [600,700,740,890,680,980,800]
standard_sodium2 = 800



# 주의 ColumnDataSource안에 들어가는 dictionary는 각 key의 value 길이가 같아야 한다.
# 길이가 다르면 가장 최소 길이의 단위로 각 데이터를 읽어온다.
data2 ={
    "days":day2,
    "cosine_simil":cosine_simil2,
    "sodium":sodium2
}



source2 = ColumnDataSource(data=data2)


# tooltip의 font_size 및 기타 사항을 직접 수정하려면 이렇게 html 코드로 작성을 해서 입력해야한다.
# 아래 코드 작성을 위한 참고 정보는 다음 링크에 있다.
# https://docs.bokeh.org/en/1.3.4/docs/user_guide/tools.html#basic-tooltips

TOOLTIPS2 = """
    <div>
        <div>
            <span style="font-size: 14px; font-weight: bold;">나트륨: @sodium mg</span>
        </div>
    </div>
"""



p2 = figure(plot_width=600, plot_height=600, x_range = (0,1100), y_range = (94,100),tools="hover"
           ,toolbar_location= None, tooltips=TOOLTIPS2, title="나트륨 정보 (일주일 기준)")
p2.title.text_font_size = '14pt'

# 범례를 그리기 위해 안보이는 점들을 찍어놓는다(안보이게 하기 위해서 축에다 찍는다. 
p2.circle(x = 200, y = 0, color = '#ACFFFB', fill_alpha = 1)
p2.circle(x = 600, y = 0, color = '#A8FF9E', fill_alpha = 1)
p2.circle(x = 900, y = 0, color = '#FFAAAA', fill_alpha = 1)

p2.scatter(x='sodium', y='cosine_simil', size=15,color = "#6134FF", source=source2)




# 산점도의 각 점위에 label 붙여주기
labels2 = LabelSet(x='sodium', y='cosine_simil', text='days', level='glyph',text_font_size="12pt", text_color="#07594A",
              x_offset=5, y_offset=5, source=source2, render_mode='canvas')

p2.add_layout(labels2)


# 산점도의 각 영역을 박스화하기
p2.add_layout(BoxAnnotation(left=0,right=400, fill_alpha=0.15, fill_color='#2EC1FF', line_color='red',line_alpha = 0))
p2.add_layout(BoxAnnotation(left=400,right=800, fill_alpha=0.15, fill_color='#00CB00', line_color='red',line_alpha = 0))
p2.add_layout(BoxAnnotation(left=800, fill_alpha=0.15, fill_color='#FC2E0D', line_color='red',line_alpha = 0))



# 범례를 만들어서 그려주기(LegendItem(범례항목 이름, 해당 범례 대상)
# p.renderers[]는 p라는 도화지에 그려진 항목들을 순서대로 호출하는 방법이다
li1_2 = LegendItem(label='안전구간', renderers=[p2.renderers[0]])
li2_2 = LegendItem(label='평균구간', renderers=[p2.renderers[1]])
li3_2 = LegendItem(label='위험구간', renderers=[p2.renderers[2]])
legend2 = Legend(items=[li1_2, li2_2, li3_2], location='top_left',background_fill_alpha = 0.7)
p2.add_layout(legend2)




# x축과 y축에 라벨 넣어주기
p2.xaxis.axis_label = '나트륨 수치 (mg)'
p2.yaxis.axis_label = '균형 식단 지수'


# show the results
show(p2)
```
